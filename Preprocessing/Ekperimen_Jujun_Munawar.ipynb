```python
X = df.drop(columns = ['diagnosed_diabetes', 'diabetes_stage'])
Y = df.diagnosed_diabetes

X_df, X_test, Y_df, Y_test = train_test_split(X,Y, test_size=0.1, random_state=42,stratify=Y)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X,Y, test_size=0.2, random_state=42,stratify=Y)
```

```python
X_train = X_train[[
    'diabetes_risk_score'
    ,'hba1c'
    ,'glucose_postprandial'
    ,'glucose_fasting'
    ,'family_history_diabetes'
    ,'hypertension_history'
    ,'cardiovascular_history'
]]
X_val = X_val[['diabetes_risk_score',
        'hba1c',
        'glucose_postprandial',
        'glucose_fasting',
       'family_history_diabetes',
       'hypertension_history',
       'cardiovascular_history']]

Y_train = Y_train.values
Y_val = Y_val.values
relevant_categorical = ['family_history_diabetes','hypertension_history','cardiovascular_history']
relevant_numeric = ['diabetes_risk_score', 'hba1c', 'glucose_postprandial', 'glucose_fasting']
```

```python
#standarisasi
scaler = StandardScaler()

X_train_categorical = X_train[relevant_categorical].values
X_val_categorical = X_val[relevant_categorical].values

X_train_numeric = X_train[relevant_numeric]
X_val_numeric = X_val[relevant_numeric]


scaler.fit(X_train_numeric)
X_train_scaled = scaler.transform(X_train_numeric)
X_val_scaled = scaler.transform(X_val_numeric)

X_train_processed = np.concatenate((X_train_scaled, X_train_categorical), axis = 1)
X_val_processed = np.concatenate((X_val_scaled, X_val_categorical), axis = 1)
```

```python
models = [LogisticRegression(class_weight='balanced'),
          RandomForestClassifier(class_weight='balanced'),
          KNeighborsClassifier(),
          SVC(class_weight='balanced')
         ]
model_name = [type(model).__name__ for model in models]
accuracy = []
training_times = []
fitted_models = []
for i, model in enumerate(models):
    start = time.process_time()
    fitted_model = model.fit(X_train_processed, Y_train)
    fitted_models.append(fitted_model)
    end = time.process_time()
    y_pred = fitted_model.predict(X_val_processed)
    print('model ', model_name[i],' fitted')
    accuracy.append(accuracy_score(Y_val, y_pred))
    training_times.append(np.round(end-start, 4))
```

```python
models_result = pd.DataFrame(
    {'models':model_name,
    'accuracy':accuracy,
     'training_time':training_times
    }
)

display(models_result)

model_result_accuracy_sorted = models_result.sort_values(by=['accuracy'], ascending=False)
model_result_time_sorted = models_result.sort_values(by=['training_time'],ascending=False)
```

```python
X_test_categorical = X_test[relevant_categorical].values
X_test_numeric = X_test[relevant_numeric]

X_test_scaled = scaler.transform(X_test_numeric)

X_test_processed = np.concatenate((X_test_scaled, X_test_categorical), axis = 1)

model = fitted_models[1] ###Random Forest

y_pred = model.predict(X_test_processed)

print('Accuarcy: ', accuracy_score(Y_test, y_pred))
print('Recall: ', recall_score(Y_test, y_pred))
print('Precision: ',precision_score(Y_test, y_pred))
fpr, tpr, thresholds = roc_curve(Y_test, y_pred)
roc_auc = auc(fpr, tpr)
print('Auc: ',roc_auc)
```
